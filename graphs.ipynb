```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import networkx as nx
```

```python
adj_liste = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}
```

```python
G = nx.from_dict_of_lists(adj_liste, create_using=nx.Graph)
```

```python
pos = nx.spring_layout(G)
```

```python
plt.figure(figsize=(8, 6))
nx.draw(G, pos, 
        with_labels=True, 
        node_color='lightblue', 
        edge_color='gray', 
        node_size=2000, 
        font_size=15)
plt.title("Graph from adjacency List")
plt.show()
```
