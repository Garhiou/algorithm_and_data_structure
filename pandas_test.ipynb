```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'Positive': [59, 15],
    'Negative': [1, 125]
}, index=['SF', 'NSF']
)
```

```python
print(df)
```

```python
tp = df.loc['SF', 'Positive']
print('True Positive: ', tp)
fp = df.loc['NSF', 'Positive']
print('False Positive: ', fp)
tn = df.loc['NSF', 'Negative']
print('True Negative: ', tn)
fn = df.loc['SF', 'Negative']
print('False Negative: ', fn)
```

```python
#Precision = TP / (TP + FP)
presicion = tp / (tp + fp)
print('Presicion: ', presicion)
```

```python
#Recall = TP / (TP + FN)
recall = tp / (tp + fn)
print('Recall: ', recall)
```
